```python
def pdf(X,mu,sigma):
    a=1/np.sqrt(2*np.pi*sigma**2)
    b=-1/(sigma**2)
    return a*np.exp(b*(X-mu)**2)

X=np.linspace(-6,6,1000)

for i in range(5):
    samples=np.random.randn(50)
    mu=np.mean(samples)
    sigma=np.std(samples)
    plt.plot(pdf(X,mu,sigma),color='.75')

plt.plot(pdf(X,0,1),color='k')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig,(ax1,ax2,ax3)=plt.subplots(1,3)
fig.set_size_inches(18.5, 4.5)
x=[1,2,3,4,5]
ax1.plot(x,[1,2,3,4,6],linestyle="dashed", marker="o", color="green")
ax1.set_title('First arm',fontsize=15)
ax1.set_xlabel('Set of states $X_t$',fontsize=15)
ax1.set_ylabel('$g_1$',fontsize=15)
ax2.plot(x,[2,4,2,4,4],linestyle="dashed", marker="o", color="green")
ax2.set_title('Second arm',fontsize=15)
ax2.set_xlabel('Set of states $X_t$',fontsize=15)
ax2.set_ylabel('$g_2$',fontsize=15)
ax3.plot(x,[3,5,5,1,2],linestyle="dashed", marker="o", color="green")
ax3.set_title('Third arm',fontsize=15)
ax3.set_xlabel('Set of states $X_t$',fontsize=15)
ax3.set_ylabel('$g_3$',fontsize=15)
```
