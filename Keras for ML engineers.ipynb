```python
dataset=numpy.loadtxt('pima-indians-diabetes.csv',delimiter=',')
```

```python
X=dataset[:,0:8]
y=dataset[:,8]
```

```python
model=Sequential()
```

```python
model.add(Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,nb_epoch=150,batch_size=10)
```

```python
scores=model.evaluate(X,y)
```

```python
print('%s:%.2f%%'%(model.metrics_names[1],scores[1]*100))
```

```python
model.fit(X,y,validation_split=0.33,epochs=150,batch_size=10)
```

```python
scores=model.evaluate(X,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.33,random_state=seed)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=150,batch_size=10)
```

```python
from sklearn.model_selection import StratifiedKFold
kfold=StratifiedKFold(n_splits=10,shuffle=True,random_state=seed)
```

```python
cvscores=[]
for train,test in kfold.split(X,y):
    model=Sequential()
    model.add(Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
    model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
    model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    model.fit(X[train],y[train],epochs=150,batch_size=10)
    score=model.evaluate(X[test],y[test])
    cvscores.append(score[1]*100)
```

```python
print('mean_acc:%.2f%%, std:%.2f%%'%(numpy.mean(cvscores),numpy.std(cvscores)))
```

```python
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
seed=7
numpy.random.seed(seed)
dataframe=pandas.read_csv('iris.csv')
dataset=dataframe.values
X=dataset[:,0:4].astype(float)
y=dataset[:,4]
```

```python
encoder=LabelEncoder()
encoded_y=encoder.fit_transform(y)
```

```python
encoded_y
```

```python
def baseline_model():
    model=Sequential()
    model.add(Dense(8,input_dim=4,activation='relu'))
    model.add(Dense(3,activation='softmax'))
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
estimator=KerasClassifier(build_fn=baseline_model,epochs=200,batch_size=5)
```

```python
kfold=KFold(n_splits=10,shuffle=True)
```

```python
results=cross_val_score(estimator,X,encoded_y,cv=kfold)
```

```python
results.mean()*100
```

```python
results.std()*100
```

```python
dataframe=pandas.read_csv('sonar.csv',header=None)
```

```python
dataset
```

```python
X=dataset[1:,:60].astype(float)
```

```python
y=dataset[1:,60]
```

```python
encoder=LabelEncoder()
encoded_y=encoder.fit_transform(y)
```

```python
def create_baseline():
    model=Sequential()
    model.add(Dense(60,input_dim=60,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
estimator=KerasClassifier(build_fn=create_baseline,epochs=100,batch_size=5)
kfold=StratifiedKFold(n_splits=10,shuffle=True)
```

```python
results=cross_val_score(estimator,X,encoded_y,cv=kfold)
```

```python
results.mean()*100
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
def create_smaller():
    model=Sequential()
    model.add(Dense(30,input_dim=60,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
estimators=[]
```

```python
estimators.append(('Standardize',StandardScaler()))
```

```python
estimators.append(('mlp',KerasClassifier(build_fn=create_smaller,epochs=100,batch_size=5)))
```

```python
pipeline=Pipeline(estimators)
```

```python
results=cross_val_score(pipeline,X,encoded_y,cv=kfold)
```

```python
print('Smaller network: %.2f%%(%.2f%%)'%(results.mean()*100,results.std()*100))
```

```python
def create_larger():
    model=Sequential()
    model.add(Dense(60,input_dim=60,activation='relu'))
    model.add(Dense(30,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
estimators=[]
estimators.append(('Standardize',StandardScaler()))
estimators.append(('mlp',KerasClassifier(build_fn=create_larger,epochs=100,batch_size=5)))
pipeline=Pipeline(estimators)
```

```python
results=cross_val_score(pipeline,X,encoded_y,cv=kfold)
```

```python
print('Larger network: %.2f%%(%.2f%%)'%(results.mean()*100,results.std()*100))
```

```python
from sklearn.datasets import load_boston
```

```python
import pandas as pd
boston = load_boston()
data = pd.DataFrame(boston.data)
data.columns = boston.feature_names
data['PRICE'] = boston.target
X, y = data.iloc[:,:-1],data.iloc[:,-1]
```

```python
def baseline_model():
    model=Sequential()
    model.add(Dense(13,input_dim=13,kernel_initializer='normal',activation='relu'))
    model.add(Dense(1,kernel_initializer='normal'))
    model.compile(loss='mean_squared_error',optimizer='adam')
    return model
```

```python
seed=7
numpy.random.seed(seed)
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
estimator=KerasRegressor(build_fn=baseline_model,epochs=100,batch_size=5)
```

```python
from sklearn.model_selection import KFold
kfold=KFold(n_splits=10,shuffle=True,random_state=seed)
```

```python
results=cross_val_score(estimator,X,y,cv=kfold)
```

```python
estimators=[]
estimators.append(('Standardize',StandardScaler()))
estimators.append(('mlp',KerasRegressor(build_fn=baseline_model,epochs=100,batch_size=5)))
pipeline=Pipeline(estimators)
```

```python
results=cross_val_score(estimator,X,y,cv=kfold)
```

```python
def larger_model():
    model=Sequential()
    model.add(Dense(13,input_dim=13,kernel_initializer='normal',activation='relu'))
    model.add(Dense(6,kernel_initializer='normal',activation='relu'))
    model.add(Dense(1,kernel_initializer='normal'))
    model.compile(loss='mean_squared_error',optimizer='adam')
    return model
```

```python
estimators=[]
estimators.append(('Standardize',StandardScaler()))
estimators.append(('mlp',KerasRegressor(build_fn=larger_model,epochs=100,batch_size=5)))
pipeline=Pipeline(estimators)
```

```python
results=cross_val_score(estimator,X,y,cv=kfold)
```
